```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("road_lanes.png")
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
_,global_thresholded = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray_img,cmap="grey")
plt.subplot(1,2,2)
plt.title("Global Threshold Image")
plt.imshow(global_thresholded,cmap="grey")
plt.show()
```

```python
adaptive_thresholded = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray_img,cmap="grey")
plt.subplot(1,2,2)
plt.title("Adaptive threshold")
plt.imshow(adaptive_thresholded,cmap="grey")
plt.show()
```

```python
_,otsu_thresholded = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray_img,cmap="grey")
plt.subplot(1,2,2)
plt.title("Otsu Threshold Image")
plt.imshow(otsu_thresholded,cmap="grey")
plt.show()
```
